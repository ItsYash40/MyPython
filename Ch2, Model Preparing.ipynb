```python
import matplotlib.pyplot as plt
#2D
X_axis= range(-20,21)
func= [x**2 for x in X_axis]
plt.plot(X_axis, func , 'g', linestyle='dashdot')
plt.xlabel('Xvalues')
plt.ylabel('Function values')
plt.title('Parabolic Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Sin Graph

X=np.linspace(0,2*np.pi,100)
sin= np.sin(X)
cos=np.cos(X)
plt.plot(X,sin, 'g', label='sin(X)')
plt.plot(X,cos, 'r', label='cos(X)')
plt.xlabel('X axis points', fontweight='bold')
plt.ylabel('Function axis points', fontweight='bold')
plt.title('Plot of SinX and CosX')
plt.legend(loc=9)
plt.show()
```
