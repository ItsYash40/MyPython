```python
import pandas as pd

data = {'grade': ['A', 'B', 'C', 'D', 'E', 'F']}
df = pd.DataFrame(data)

grade_map = {
    "A": "90-100",
    "B": "80-89",
    "C": "70-79",
    "D": "60-69",
    "E": "50-59",
    "F": "40-49"
}

df["num_update"] = df["grade"].map(grade_map)
print(df)
```

```python
# Binning Continious variable

import numpy as np
import pandas as pd

df = pd.DataFrame(
    {'Price': [20000, 30000, 12000, 60000,-1]}
)

bins = [0, 15000, 40000, np.inf]   
labels = ['Low', 'Medium', 'High']

df["Price_Category"] = pd.cut(df["Price"], bins=bins, labels=labels)
print(df)
```

```python
# Binning Continious  (Review & Rating)

import numpy as np
import pandas as pd

df = pd.DataFrame(
    {'Review': [20000, 30000, 12000, 60000,-1,25,1394861,95]}
)

bins = [0, 15000, 40000, np.inf]   
labels = ['*', '**', '***']

df["Rating"] = pd.cut(df["Review"], bins=bins, labels=labels)
print(df)
```

```python
# LDA

import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X, y = iris.data, iris.target

# Apply LDA
lda = LinearDiscriminantAnalysis(n_components=2) 
X_lda = lda.fit_transform(X, y)

# Print shapes
print("Original shape:", X.shape)
print("After LDA:", X_lda.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load iris dataset
iris = datasets.load_iris()
predictors = iris.data  
target = iris.target
predictors = StandardScaler().fit_transform(predictors)

# Apply PCA
pca = PCA(n_components=2)
p1_C = pca.fit_transform(predictors)

# Create a DataFrame 
p1_C_ds = pd.DataFrame(data=p1_C, columns=['PC-1', 'PC-2'])
print(p1_C_ds.head())

# ==== PCA result ====

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(p1_C[:, 0], p1_C[:, 1], c=target, cmap='viridis', edgecolor='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Iris Dataset')
plt.show()


# Box Plot
plt.figure(figsize=(8,6))
plt.boxplot([p1_C[:, 0], p1_C[:, 1]], labels=['PC-1', 'PC-2'])
plt.title("Box Plots of Principal Components")
plt.ylabel("Value")
plt.show()
```
