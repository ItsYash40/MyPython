```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8')
x=np.linspace(0,2*np.pi, 100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1, 'r', label='sin(x)')
plt.plot(x,y2, 'g', label='cos(x)')
plt.xlevel('x points of cosx')
plt.ylevel('y points of cosy')

plt.show()
```

```python
fig,ax=plt.subplots(2,1,figsize=(8,8))

ax[0].plot(x,y2,'r')
ax[1].plot(x,y2,'c')
ax[0].set_ylabel('cos(x)')
ax[1].set_ylabel('cos(2x)')
ax[0].set_xlabel('x points of cosx')
ax[1].set_ylabel('y points of cos 2x')
fig.suptitle('Graph')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

l1=['a','b','c','d','e']
l2=['8','10','11','17','21']

colors=['r','g','b','orange','violet']
colors=np.random.rand(len(l2),3)
plt.bar(l1,l2,color=colors,edgecolor="black", linewidth=1.4)
plt.title("My list")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

l1 = ['a', 'b', 'c', 'd', 'e']
l2 = ['8', '10', '11', '17', '21']

l2 = list(map(int, l2))
colors = np.random.rand(len(l2), 3) 

plt.pie(l2, labels=l1, colors=colors,autopct='%1.1f%%' ,shadow=True, wedgeprops={'edgecolor': 'black', 'linewidth': 1.4}, startangle=90)
plt.title("My list")
plt.show()
```
